## Deliverable 2: Scrape and Analyze Mars Weather Data

```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
```

```python
browser = Browser('chrome')
```

### Step 1: Visit the Website

Use automated browsing to visit the [Mars Temperature Data Site](https://static.bc-edx.com/data/web/mars_facts/temperature.html). Inspect the page to identify which elements to scrape.

   > **Hint** To identify which elements to scrape, you might want to inspect the page by using Chrome DevTools to discover whether the table contains usable classes.


```python
# Visit the website
# https://static.bc-edx.com/data/web/mars_facts/temperature.html
url = "https://static.bc-edx.com/data/web/mars_facts/temperature.html"
browser.visit(url)
```

### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Create a Beautiful Soup object
soup = BeautifulSoup(browser.html, 'html.parser')

# Extract the table from the HTML
table = soup.find('table')
```

```python
# Extract table rows
rows = table.find_all('tr')
```

### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Here’s an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
# Create an empty list to hold rows of data
data = []
# Extract headers th= table headers
headers = [th.text for th in table.find_all('th')]
# Extract each row's data
for row in rows[1:]:  # Skip the header row
    cols = row.find_all('td') # td = table data
    cols = [col.text.strip() for col in cols]
    data.append(dict(zip(headers, cols)))
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
df = pd.DataFrame(data)
```

```python
# Confirm DataFrame was created successfully
df.head()
```

### Step 4: Prepare Data for Analysis

Examine the data types that are currently associated with each column. If necessary, cast (or convert) the data to the appropriate `datetime`, `int`, or `float` data types.

  > **Hint** You can use the Pandas `astype` and `to_datetime` methods to accomplish this task.


```python
# Examine data type of each column
df.dtypes
```

```python
# Change data types for data analysis
df['terrestrial_date'] = pd.to_datetime(df['terrestrial_date'])
df['sol'] = df['sol'].astype(int)
df['ls'] = df['ls'].astype(int)
df['month'] = df['month'].astype(int)
df['min_temp'] = df['min_temp'].astype(float)
df['pressure'] = df['pressure'].astype(float)
```

```python
# Confirm type changes were successful by examining data types again
df.dtypes
```

### Step 5: Analyze the Data

Analyze your dataset by using Pandas functions to answer the following questions:

1. How many months exist on Mars?
2. How many Martian (and not Earth) days worth of data exist in the scraped dataset?
3. What are the coldest and the warmest months on Mars (at the location of Curiosity)? To answer this question:
    * Find the average the minimum daily temperature for all of the months.
    * Plot the results as a bar chart.
4. Which months have the lowest and the highest atmospheric pressure on Mars? To answer this question:
    * Find the average the daily atmospheric pressure of all the months.
    * Plot the results as a bar chart.
5. About how many terrestrial (Earth) days exist in a Martian year? To answer this question:
    * Consider how many days elapse on Earth in the time that Mars circles the Sun once.
    * Visually estimate the result by plotting the daily minimum temperature.


```python
# 1. How many months are there on Mars?
months = df['month'].nunique()
months
```

```python
# 2. How many sols (Martian days) worth of data are there?
sols = df['sol'].nunique()
sols
```

```python
# 3. What is the average minimum temperature by month?
avg_temp_by_month = df.groupby('month')['min_temp'].mean()
avg_temp_by_month
```

```python
# Plot the average minimum temperature by month
avg_temp_by_month.plot(kind='bar', title='Average Minimum Temperature by Month')
plt.xlabel('Month')
plt.ylabel('Average Min Temperature (°C)')
plt.show()
```

```python
# Identify the coldest and hottest months in Curiosity's location by sorting the previous graph
avg_temp_by_month_df = avg_temp_by_month.reset_index()
avg_temp_by_month_sorted = avg_temp_by_month_df.sort_values(by='min_temp')

# Plotting using Matplotlib
plt.bar(avg_temp_by_month_sorted['month'].astype(str), avg_temp_by_month_sorted['min_temp'])

# Adding labels and title
plt.title('Average Minimum Temperature by Month')
plt.xlabel('Month')
plt.ylabel('Average Min Temperature (°C)')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# 4. What is the average pressure by month?
avg_press_by_month = df.groupby('month')['pressure'].mean()
avg_press_by_month
```

```python
# Plot the average pressure by month
avg_press_by_month.plot(kind='bar', title='Average Atm. Pressure by Month')
plt.xlabel('Month')
plt.ylabel('Atmospheric Pressure')
plt.show()
```

```python
# Identify the lowest and highest pressure months in Curiosity's location by sorting the previous graph
avg_press_by_month_df = avg_press_by_month.reset_index()
avg_press_by_month_sorted= avg_press_by_month_df.sort_values(by='pressure')

# Plotting using Matplotlib
plt.bar(avg_press_by_month_sorted['month'].astype(str), avg_press_by_month_sorted['pressure'])

plt.title('Average Atm. Pressure by Month')
plt.xlabel('Month')
plt.ylabel('Atmospheric Pressure')
plt.show()
```

```python
# 5. How many terrestrial (Earth) days are there in a Martian year?
# Visually estimate the result by plotting the daily minimum temperature of each observation in the data set.

plt.plot(df['sol'], df['min_temp'])
plt.title('Daily Minimum Temperature on Mars')
plt.xlabel('Number of terrestrial days')
plt.ylabel('Minimum Temperature (°C)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### Minimum Temperature

- The average minimum temperatures on Mars show a range from about -83.31°C in March to -68.38°C in August. This range highlights the extreme cold temperatures typical of Mars, with fluctuations across the Martian year.
- The coldest months are observed in the early months of the Martian year (months 3 and 4), with average temperatures around -83°C to -82.75°C. This suggests a period of colder temperatures, which could correlate with the Martian winter.
- The colder months correspond with the Martian winter, while the warmer months correspond with Martian summer. This cyclical pattern is crucial for planning further missions.

#### Atmospheric Pressure

- The average atmospheric pressure ranges from 745.05 Pa in JUN to 913.31 Pa in SEP.
- The lowest average pressure is recorded in JUNE, corresponding to the peak of summer.
- The highest average pressure is observed in SEP (913.31 Pa). This is likely associated with the Martian winter or autumn.

```python
# Year length calculation

# To datetime formatting
df['terrestrial_date'] = pd.to_datetime(df['terrestrial_date'])

# We'll use the first observed year for taking the month data
first_year = df[(df['terrestrial_date'] >= '2013-08-01') & (df['terrestrial_date'] <= '2015-06-18')]

# Group the filtered data by month and find the first and last date for each month
month_ranges = first_year.groupby('month')['terrestrial_date'].agg(['min', 'max'])

# Calculate the duration of each month in Earth days
month_ranges['duration_in_days'] = (month_ranges['max'] - month_ranges['min']).dt.days

# Calculate the year duration
year_days= month_ranges['duration_in_days'].sum()

print(f'The Martian year length: {year_days} Earth days')
```

#### Year Length

- Data Observation (last plot): Patterns in minimum temperatures around days 500 and 1250 suggest a Martian year length near 750 Earth days.
- Estimate Refinement: From data spanning August 1, 2013, to June 18, 2015, we approximated the Martian year to be about 675 Earth days.
- Validation: The official Martian year length is 687 Earth days. We can confirming our approach is reasonably accurate.
